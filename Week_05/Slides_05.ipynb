# Vector field

```octave
[x, y] = meshgrid (-3:0.2:3);
h = quiver(x, y, x, 2*y);
title('vector field (x 2y)')
```

```octave
colormap(jet)
ticks = [-3.0:0.1:3.0];
[x, y] = meshgrid(ticks, ticks);
z = x.^2/2 + y.^2;
surf(x,y,z);
title('Elliptic Bowle using meshgrid')
```

```octave
colormap(jet)
ticks = [-3.0:0.1:3.0];
[x, y] = meshgrid(ticks, ticks);
z = x.^2 - y.^2;
surf(x,y,z);
title('Elliptic Bowle using meshgrid')
```

## Countour Line
