# 11 Plotting with Octave

$y = sinx$ in the interval $[0,2Ï€]$.

```octave
x = linspace(0, 2*pi, 100);
y = sin(x);
plot(x, y, ';sin(x);');
title ('sin(x)');
grid on
xlabel ('x')
ylabel ('y')
```

## Countourplots

```octave
x = linspace(-2 * pi,2 * pi);
y = linspace(0,4 * pi);
[X,Y] = meshgrid(x,y);
Z = X.^2-Y.^2;

figure
contour(X,Y,Z)
```

## Plot Graph

```octave
[X,Y] = meshgrid(-2:.2:2, -2:.2:2);
Z = X .* exp(-X.^2 - Y.^2);
mesh(X,Y,Z)
surf(X,Y,Z)
meshc(X,Y,Z)
axis off
```
