# Clamped example on slides

Generate the data, i.e. 9 points on the curve $y = \frac{1}{1+25x^2}$ where $-1\leq x\leq 1$.

```octave
X = linspace(-1, 1, 9);
Y = 1./(1 + 25*X.^2);
```

Calculate $100$ model points and compute the not-a-knot spline

```octave
x = linspace(-1,1);
y = spline(X, [1 Y -4], x);
```

Calculate $100$ data points just for visualization

```octave
yf = 1./(1 + 25*x.^2);
```

Plot the data points (blue circles), the not-a-knot spline (solid red) and the function (dashed blue).

```octave
plot(X, Y, 'bo', 'linewidth', 4, x, y, 'r-', 'linewidth', 4, x, yf, 'b--')
set(gca, "linewidth", 4, "fontsize", 16)
```
